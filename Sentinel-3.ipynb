## S3 OLCI

```python
x1 = 11.012421
y1 = 46.659805
x2 = 11.505432
y2 = 47.009289
```

```python
bbox = x1, y1, x2, y2
```

```python
spatial_res = 0.003
time_range = ['2018-05-14', '2018-07-31']
band_names = ['B04']
```

```python
SH = SentinelHub()
SH.dataset_names
```

```python
SH.band_names('S3OLCI')
```

```python
cube_config = CubeConfig(dataset_name='S3OLCI',
                         band_names=band_names,
                         tile_size=[512, 512],
                         bbox=bbox,
                         spatial_res = spatial_res,
                         time_range = time_range,
                         time_tolerance='30M')  
```

```python
cube = open_cube(cube_config=cube_config, api_url="https://creodias.sentinel-hub.com")
```
